```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load dataset
df = pd.read_csv('simulated_recommendation_dataset.csv')

# Drop non-predictive columns
df_features = df.drop(columns=['user_id', 'problem_id', 'label'])
```

```python
encoder = OneHotEncoder(sparse_output=False)

encoded_topics = encoder.fit_transform(df_features[['problem_topic']])
encoded_topic_df = pd.DataFrame(encoded_topics, columns=encoder.get_feature_names_out(['problem_topic']))
```

```python
df_numerical = df_features.drop(columns=['problem_topic'])
X = pd.concat([df_numerical.reset_index(drop=True), encoded_topic_df.reset_index(drop=True)], axis=1)
```
